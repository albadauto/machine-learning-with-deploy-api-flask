```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df_sentiments = pd.read_csv("dataset/train.csv",index_col=False, encoding="latin-1")
```

```python
df_sentiments
```

```python
df_sentiments = df_sentiments.drop(columns=['textID', 'text','Time of Tweet','Age of User','Country', 'Population -2020', 'Land Area (Km²)', 'Density (P/Km²)'])
```

```python
df_sentiments
```

```python
vectorizer = CountVectorizer()
df_sentiments_vectorizer = vectorizer.fit_transform(df_sentiments['selected_text'].fillna(''))
```

```python
sentiments = pd.DataFrame(df_sentiments_vectorizer.toarray())
```

```python
sentiments
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_sentiments['sentiment'] = encoder.fit_transform(df_sentiments['sentiment'])
df_sentiments
```

```python
X = sentiments
y = df_sentiments['sentiment']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 5)
```

```python
X_train
```

```python
X_train.shape[1]
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units=400, activation='relu', input_shape=(17610,)))
model.add(tf.keras.layers.Dense(units=400, activation='relu'))
model.add(tf.keras.layers.Dense(units=400, activation='relu'))
model.add(tf.keras.layers.Dense(units=3, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy')
history = model.fit(X_train, y_train, epochs=10)
```

```python
mytest = ["Renata, i love you so much"]
mytest = vectorizer.transform(mytest)
mytest_array = mytest.toarray()
mytest_array.shape
predict = model.predict(mytest_array)
```

```python
#Neutro = 1, Negativo = 0, Positivo = 2
result = np.argmax(predict)
match result:
    case 0: 
        print("O sentimento é Negativo")
    case 1:
        print("O sentimento é Neutro")
    case 2:
        print("O sentimento é Positivo")
        
```

```python
#Salva o modelo criado e treinado em um arquivo .h5
model.save('testmodel.h5')
```

```python
savedModel = tf.keras.models.load_model('testmodel.h5')
savedModel.summary()
```
